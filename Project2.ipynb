```python
import pandas as pd 
import numpy as nd
import matplotlib.pyplot as plt
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("csv_files\\hiring.csv")
df
```

```python
import math
median_score=math.floor(df["test_score(out of 10)"].median())
median_score
```

```python
df["test_score(out of 10)"]=df["test_score(out of 10)"].fillna(median_score)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
df['experience']= df['experience'].apply(w2n.word_to_num)
x_label = df[['experience','test_score(out of 10)','interview_score(out of 10)']]
y_label = df['salary($)']
reg.fit(x_label,y_label)
```

```python
%matplotlib inline 
plt.plot(x_label,y_label,"r--.")
plt.plot(x_label,reg.predict(x_label))
```
