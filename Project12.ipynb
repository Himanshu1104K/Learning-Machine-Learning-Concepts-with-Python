```python
import pandas as pd
import numpy as nd
```

```python
df = pd.read_csv('csv_files\\titanic.csv')
df
```

```python
x = df.Age 
x
```

```python
import math
median_x = math.floor(df['Age'].median())
median_x
```

```python
df.Age=df.Age.fillna(median_x)
df
```

```python
from sklearn.preprocessing import LabelEncoder
ln = LabelEncoder()
```

```python
new_df = df[['Survived','Pclass','Sex','Age','Fare']]
new_df
```

```python
new_df['Sex_ln'] = ln.fit_transform(new_df.Sex)
```

```python
new_df = new_df.drop(['Sex'],axis='columns')
new_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(new_df.drop(['Survived'],axis='columns'),new_df.Survived,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,model.predict(x_test))
cm
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,4))
sn.heatmap(cm,annot=True)
```
