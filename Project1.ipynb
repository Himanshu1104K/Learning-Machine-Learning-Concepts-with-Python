```python
import numpy as nd 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("csv_files\\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.year,df.perCapitaIncome,color="r",marker="*")
plt.xlabel('Year')
plt.ylabel("Per Captial Income (US$)")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.perCapitaIncome)
```

```python
reg.predict([[2016]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.plot(df.year,df.perCapitaIncome,'r-')
plt.xlabel('Year')
plt.ylabel("Per Captial Income (US$)")
plt.plot(df.year,reg.predict(df[["year"]]),color="blue")
```
